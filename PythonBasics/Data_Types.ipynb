### Data Types

1. Defination
. Data types are classification of data which tell the compiler or interpreter howw the programmer intends to use the data.

. They determine the types of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.
2. Importance of Data Types in programing
. Data Types are ensure that data is stored in an efficint way.
.  They help in performing correct operation on data.
. Proper use of data types can prevent errors and bugs in the program.

```python
## Integer DataTypes
age=22
print(age)
print(type(age))
```

```python
## Float DataTypes
height=5.7
print(height)
print(type(height))
```

```python
## String datatypes example
name="Rahul"
print(name)
print(type(name))
```

```python
## boolean datatypes

is_true=True
print(type(is_true))
```

```python
## common error
result="hello" + 5
print(result)
```
