```python
## While Loop

## The while loop continues to excute as long as the condition is True.

count=0

while count<5:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## break - The break statements exits the loop permaturely

## break statements

for i in range(10):
    if i == 5:
        break
    print(i)
    
```

```python
## continue

## The continue statements skips the current iteration with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
## The pass statements is a null operation; it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loops
## a loop inside loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sum of first N natural numbers using a while and for loop

## While Loop

n=10
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1
    
print(sum)    
```

```python
## for loop

n=10
sum=0

for i in range(11):
    sum=sum+i
    
print(sum)    
    
```

```python
## Example - Prime numbers between 1 and 100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)    
```
